```python
import math
import numpy as np
import matplotlib.pyplot as plt

#D_list=[0.1, 0.5]
D_list=[0.1, 0.15, 0.2, 0.3, 0.4, 0.5, 0.7, 1]
tau_list = np.arange(0.001, 0.1, 0.001)
fig = plt.figure(figsize=(10,4))

for D in D_list:
    for tau in tau_list:
        Upp0 = 1.0   # abs(U''(0)) = 1
        Upp1 = 2.0   # U''(1) = 2
        prefactor = 2.0 * math.pi / math.sqrt(Upp0 * Upp1)  # = 2π / sqrt(2) = sqrt(2) π

        # We want 1/(1 - a tau) => a=1 => (1 - tau)
        denominator = (1.0 - tau)
        exponential = math.exp((1.0/(4.0 * D)) * (1.0 + 2.0 * 1.0 * tau))

        T_fox = prefactor / denominator * exponential

        #print('D =',D, ', tau =',tau, '\t',f"T = {T_fox:.3f}")
        plt.plot(tau, T_fox, 'o', ms = 10, color = (D, 0.5, min(1,D*5)))
    plt.plot(tau, T_fox, 'o', ms = 10, color = (D, 0.5, min(1,D*5)), label = 'D = '+ str(D))
    #plt.plot(x_list[i_range*100 - 1], y_list[i_range*100 - 1], 'o',ms = 10, color = (1, 0, 0.5), label = str(np.round(i_range*100/len(x_list), 5)) + '% of trajectory')
    #plt.plot(x, y, 'ko', ms = 10, label = 'end')
plt.legend()
plt.grid()
plt.xlabel("tau")
plt.ylabel("T_fox")
plt.show()

```
